```python
#importing and loading file
import json
def load_data(filename):
    """Loads JSON data from a given file."""
    try:
        with open(filename, 'r') as file:
            # TODO Hints: Use the json module to load the data from the 'file'
            # Hint 2: Replace 'pass' with your code and return the loaded data
            return json.load(file)

    except FileNotFoundError:
        print(f"\n Warning: '{filename}' not found. Starting with an empty dataset.")
        return []
    except json.JSONDecodeError:
        print(f"\n Error: '{filename}' is not formatted correctly. Starting empty.")
        return []
#searching and processing
def search_data(data, search_key, search_value):
    """Searches a list of dictionaries for a specific key-value pair."""
    results = []

    for record in data:
        # TODO: Check if the 'record' is a dictionary AND contains the 'search_key'
        # TODO: If it does, compare the value of the record's key with 'search_value'
        # Hint: Convert both values to strings and use .lower() to make your search case-insensitive!
        # TODO: If it's a match, append the record to the 'results' list
        if isinstance(record, dict) and search_key in record:
            if str(record.get(search_key)).lower() == str(search_value).lower():
                results.append(record)
    return results
#saving the output
def save_data(data, filename):
    """Saves a list of dictionaries to a JSON file."""
    try:
        with open(filename, 'w') as file:
            # TODO: Use the json module to dump 'data' into the 'file'
            # Hint: Pass the argument `indent=4` to make the saved JSON file readable for humans
            json.dump(data, file, indent=4)
            print(f"\n Success: Data saved to '{filename}'.")
    except Exception as e:
        print(f"\n Error saving file: {e}")
#interaction with the user
def main():
    print("Welcome to the JSON Explorer Tool!")
    filename = input("Enter the name of your JSON file (e.g., data.json): ")

    # Load the initial data
    dataset = load_data(filename)
    latest_results = [] # Keep track of search results so we can save them later

    while True:
        print("\n Main Menu ")
        print("1. Search Records")
        print("2. Save Last Search Results")
        print("3. Quit")

        choice = input("Choose an option (1-3): ")

        if choice == '1':
            key = input("What field do you want to search by? (e.g., category): ")
            value = input(f"What value are you looking for in '{key}'?: ")

            # TODO: Call your 'search_data' function using dataset, key, and value
            # Assign the returned value to the 'latest_results' variable
            # Call the search function and assign to latest_results
            latest_results = search_data(dataset, key, value)
            # This code dynamically prints the contents of the found dictionaries
            # It only runs properly if latest_results gets populated!
            if latest_results:
                print(f"\n Found {len(latest_results)} matching records:")
                for item in latest_results:
                    print("-" * 20)
                    for k, v in item.items():
                        print(f"{k}: {v}")
            else:
                print("\n No matching records found.")

        elif choice == '2':
            if not latest_results:
                print("\n You need to search for something before you can save!")
            else:
                out_file = input("Enter a name for the new save file (e.g., results.json): ")
                # TODO: Call your 'save_data' function to save 'latest_results' into 'out_file'
                save_data(latest_results, out_file)

        elif choice == '3':
            print("Goodbye! Keep coding!")
            break

        else:
            print("Invalid choice. Please enter 1, 2, or 3.")

if __name__ == "__main__":
    main()


```
